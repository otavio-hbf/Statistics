```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import random 
```

```python
def prob_of_6(n_throws):
  n_6 = 0;
  for i in range(0, n_throws):
    die_side = random.randint(1,6)
    if(die_side == 6):
      n_6 += 1

  return float(n_6)/float(n_throws)
```

```python
prob = []
quantity = []
expected = []
for i in range (1, 1001):
  prob.append(prob_of_6(i))
  quantity.append(i)
  expected.append(1/6)

d = {'quantidade': quantity, 'probabilidade': prob, 'valor esperado':expected}
df = pd.DataFrame(data=d)

fig = px.line(df, x='quantidade', y=['probabilidade', 'valor esperado'])
fig.show()
```
